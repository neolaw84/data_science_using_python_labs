## Python Data Types 

### Primitive Data Types

There are 4 primitive data types in python. They are : 

* int
* float
* str and
* bool

### Non-primitive Data Types

Primary non-primitive data types include: 

* list
* dict and 
* tuple

Primitive data types and tuples can be **hashed** or **indexed**. Therefore they can be a dictionary's keys. `list` and `dict` can't be indexed. 

```python
t = ("I", "am", "a", "hashable", "tuple")
l = ["I", "am", "a", "list"]
d = {
    "k1" : "value_1",
    2: "value_2",
    (1, 2): {
        "a": "b"
    }
}
```

#### `list` and `dict` Comprehension

```python
fruits = ["lemon", "grape", "orange", "papaya", "banana"]
prices = [10, 20.0, 10.50, 40, "sold out"]

# this is dict comprehension
fruit_2_price = {f: p for f, p in zip (fruits, prices)}

# these are list comprehension
sour_fruits = [f for f in fruits if f in ["lemon", "grape", "orange"]]
available_prices = [p for p in prices if p != "sold out"]
```


## Control Flow

### Sequential

Python program executes one statement after another in top to bottom fashion.

### Conditional

A simple if

```python
a = input("want to quit ? [Y/N] : ")
a = a.upper()
if a == 'Y':
    exit()
```

An if/else structure

```python
a = input("want to quit ? [Y/N] : ")
a = a.upper()
if a == 'Y':
    exit()
else:
    do_something()
```

An if/elif/else structure

```python
c = input("[R]ock/[P]aper/[S]cissor ? :")
c = c.upper()
if c == 'R':
    print ("you choose Rock")
elif c == 'P':
    print ("you choose Paper")
elif c == 'S':
    print ("you choose Scissor")
else:
    print ("invalid selection)
```

### Loop

There are two kinds of loops in python

#### For

```python
for f in ["lemon", "grape", "apple"]:
    print (f)
```

#### While

```python
while True:
    do_something()
    a = ""
    while a not in ["Y", "N"]:
        a = input("continue [Y/N] ? : ")
        a = a.upper()
    if a == "N":
        break
```

### Slicing and Dicing

special function `range` ကို မှတ်မိတယ်မဟုတ်လား။ အဲဒီ `range` လိုပဲ dimension တခု အတွက် index ကို colon ':' ခံပြီး ပေးလို့ရတယ်။ 

```python
A = np.array([
    [0, 1, 2, 3, 4],
    [10, 11, 12, 13, 14],
    [20, 21, 22, 23, 24],
    [30, 31, 32, 33, 34],
    [40, 41, 42, 43, 44]
] )

print (A[1:4:2]) # this will print second row, third row
print ("-------------------")
print (A[1:4:2, 2:4]) # this will print [12, 13] and [22, 23]
```

`range` လိုပဲ ပထမ (ဘယ်ဘက်က) inclusive, ဒုတိယ (အလယ်က) exclusive ဖြစ်တယ်။ နောက်ဆုံး တလုံးက step ဖြစ်တယ်။

**Updating the slices**

```python
A[1:4:2, 2:4] = [
    [112, 113],
    [132, 133]
]
print (A) # what do you see ?
```

## `numpy` is more efficient than `list`

`numpy` uses a technique called **vectorizing** to speed up array processing. 

Also, `numpy` has stricter **data typing**. A `list` can contain all sorts of data. An `array` can't.

Try it yourself.

**Quiz** : What type of jupyter magic do you use to measure run-time of a cell ?

````python
import random
import numpy as np

random.seed(42)
n = 100
my_list = [random.randint(0, n - 1) for i in range(0, 1000)]
my_np_array = np.array(my_list)
````

````python
def sum_of_square(list):
    return sum([x**2 for x in list])

def sum_of_square_np(arr):
    return np.sum(np.square(arr))
````

````python
%%timeit -n1
sum_of_square(my_list)
````

````python
%%timeit -n1
sum_of_square_np(my_np_array)
````
