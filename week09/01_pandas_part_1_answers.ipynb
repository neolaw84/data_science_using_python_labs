````python
# write a function, when given a name "Last Name, Honorific. First Name", return Last Name, Honorific and First Name
def divide_name(name):
    # write code here to divide name to last_name, honorific and first_name
    # first divide it into last_name and honorific
    ln_n_rest = name.split(",")
    last_name = ln_n_rest[0].strip()
    # re-join all others (if any)
    rest = " ".join([x.strip() for x in ln_n_rest[1:]])
    # then split using the dot
    h_n_fn = rest.split(".")
    honorific = h_n_fn[0].strip()
    first_name = " ".join([x.strip() for x in h_n_fn[1:]])
    return last_name, honorific, first_name

print (divide_name("Graham, Miss. Margaret Edith"))
print (divide_name('Johnston, Miss. Catherine Helen "Carrie"'))
````

### အထူး function `apply`

အရင်တွေ့ခဲ့ဖူးတဲ့ `map` function လိုပဲ `apply` function ရှိတယ်။ သူ့အလုပ်လုပ်ပုံက default `axis=0` မှာ column by column, `axis=1` ဆိုရင် row by row အလုပ်လုပ်တယ်။

````python
df_titanic[["last_name", "honorific", "first_name"]] = df_titanic.apply(lambda x : divide_name(x.Name), axis=1, result_type="expand")
df_titanic
````

````python
# write code to add a column "embarked_full" based on value of "embarked" column
# C = Cherbourg, Q = Queenstown, S = Southampton
short_to_full_name = {
    "C" : "Cherbourg",
    "Q" : "Queenstown", 
    "S" : "Southampton"
}
def full_name(short_name):
    if short_name in short_to_full_name.keys():
        return short_to_full_name[short_name]
    else:
        return "not in dictionary"
df_titanic["embarked_full"] = df_titanic.apply(lambda x : full_name(x.Embarked), axis=1)
df_titanic.head()
````

````python
# based on the previous DataFrame, write code to obtain a DataFrame with 
# PassengerId, honorific_Mr, honorific_Ms, honorific_Mrs, honorific_Others with honorific_<x> having boolean values

# first, let's see what values honorific column has
df_titanic.honorific.unique()
````

````python
df_sub["honorific_Mr"] = df_titanic.honorific == "Mr"
df_sub["honorific_Mrs"] = df_titanic.honorific == "Mrs"
df_sub["honorific_Ms"] = df_titanic.honorific == "Miss"
df_sub["honorific_Others"] = ~(df_titanic.honorific.isin(["Mr", "Mrs", "Miss"]))
df_sub
````

````python
# based on the previous DataFrame, write code to obtain a DataFrame with
# PassengerId, Fare_25, Fare_50, Fare_75 and Fare_100 (they will have boolean values based on the percentile of fare they paid)

# let's get a data series of fare_information
ds_fare_info = df_titanic.Fare.describe()
ds_fare_info
````

````python
ds_fare_info["75%"]
````

````python
df_sub_2 = df_titanic[["PassengerId"]]
df_sub_2["Fare_25"] = df_titanic.Fare <= ds_fare_info["25%"]

# the following needs to get rid of Fare_25 == True
df_sub_2["Fare_50"] = ~(df_sub_2.Fare_25) & (df_titanic.Fare <= ds_fare_info["50%"] )

# same here
df_sub_2["Fare_75"] = ~(df_sub_2.Fare_25) & ~(df_sub_2.Fare_50) & (df_titanic.Fare <= ds_fare_info["75%"] )

df_sub_2["Fare_100"] = df_titanic.Fare > ds_fare_info["75%"] 
df_sub_2
````
